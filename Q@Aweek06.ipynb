```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_excel('Dataset_w6.xlsx')
```

```python
df.head(10)
```

```python
#find out how many countries are there in dataset.
print("Number of countries in the dataset:",df.country.count())
```

```python
#examine data by using visualization.
fig, ax = plt.subplots(figsize=(10, 6))
ax.scatter(x = df['Population in thousands (2017)'], y = df['Agricultural production index '],c='green', s=75,alpha=0.5)
plt.xlabel('Population in thousands (2017)')
plt.ylabel("Agricultural production index ")
plt.title("Population in thousands to Agricultural production index ")
plt.show()
```

```python
plt.figure(figsize=(30,22))
plt.title("Top 5 Countries w.r.t to International trade: Imports (million US$) ",fontsize=26)
sns.barplot(x='country', y='International trade: Imports (million US$)', data=df.nlargest(5,'Population in thousands (2017)'))
```

```python
plt.figure(figsize=(14,10))
plt.title("Distribution of GDP in various Region")
sns.barplot(x='Region', y='GDP per capita (current US$)', data = df)
plt.xticks(rotation=45)
```

```python
#Find out how many countries with respect to the region are available in the dataset
df['Region'].value_counts()
```

```python
# On average which region has the highest
```

```python
#a. GDP per capita (current US$)
high_dgp = df.groupby('Region')['GDP per capita (current US$)'].mean() 
high_dgp.nlargest(1)
```

```python
#b. International trade: Imports (million US$)
high_imports = df.groupby('Region')['International trade: Imports (million US$)'].mean() 
high_imports.nlargest(1)
```

```python
#Find out which region has more consistent with respect to GDP per capita (current US$)
std1 = df.groupby('Region')['GDP per capita (current US$)'].std() 
print("Region has more consistent with respect to GDP percapita (current US$) is:")
std1.nsmallest(1)
```
